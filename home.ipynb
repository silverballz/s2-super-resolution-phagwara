```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Load bands
red = rasterio.open("/Users/anuragsharma/Documents/S2_L2A_Phagwara_RGBN/T43REQ_20250602T054251_B04_10m.jp2").read(1).astype('float32')
nir = rasterio.open("/Users/anuragsharma/Documents/S2_L2A_Phagwara_RGBN/T43REQ_20250602T054251_B08_10m.jp2").read(1).astype('float32')

# Compute NDVI
ndvi = (nir - red) / (nir + red + 1e-6)

# Plot NDVI
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar(label="NDVI")
plt.title("NDVI Preview – Sentinel-2 (Phagwara)")
plt.show()
```
